#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

`La casa más grande está ubicada en Calle Gran Vía, Nº38 y su superficie es de 5000 metros`

Este ejercicio es similar al anterior, pero buscamos las casas más grandes y pequeñas en base a su superficie.

```python
# TODO
import pandas as pd
# Cargar el dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')
# Casa más grande
casa_mas_grande = ds[ds['surface'] == ds['surface'].max()]
# Casa más pequeña
casa_mas_pequena = ds[ds['surface'] == ds['surface'].min()]

# Imprimir resultados
print(f"La casa más grande está ubicada en {casa_mas_grande['customZone'].values[0]} y su superficie es de {casa_mas_grande['surface'].values[0]} metros")
print(f"La casa más pequeña está ubicada en {casa_mas_pequena['customZone'].values[0]} y su superficie es de {casa_mas_pequena['surface'].values[0]} metros")
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
# TODO
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Filtrar por población
arroyomolinos = ds[ds['level5'] == "Arroyomolinos (Madrid)"]

# Calcular la media de precios
media_precio = arroyomolinos['price'].mean()

# Imprimir resultado
print(f"La media de precios en Arroyomolinos (Madrid) es: {media_precio}")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt
import pandas as pd

# Trazar el histograma de precios
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Filtrar por población
arroyomolinos = ds[ds['level5'] == "Arroyomolinos (Madrid)"]

# Trazar histograma
plt.figure(figsize=(10, 6))
plt.hist(arroyomolinos['price'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribución de precios en Arroyomolinos (Madrid)')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

## 
```
